#### ReLU function

```python
def relu(input):
    '''Define your relu activation function here'''
    # Calculate the value for the output of the relu function: output
    output = max(input, 0)
    
    # Return the value just calculated
    return(output)
```


#### Input (Account 3, Children 5)

```python
import numpy as np
input_data = np.array([3,5])
input_data
```

### Weight layer 1

```python
weights_I_H1 = np.array([[2,4],
                         [4,-5]])
    
weights_I_H1
```

### Calculate: Hidden layer 1

```python
hidden_layer_input = input_data.dot(weights_I_H1) 
print(hidden_layer_input)
```

### Apply Activ. Func. : Hidden Layer 1

```python
f =  np.vectorize(relu)
hidden_layer_output = f(hidden_layer_input)
print(hidden_layer_output)
```
