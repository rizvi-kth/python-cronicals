### Define text processing function

```python
def process_document(single_newsItem_tokens):

    # data[0]['Tokens']
    # single_newsItem_tokens = data[0]['Tokens']

    # Remove BERT tokens
    cln_tokens = [t for t in single_newsItem_tokens if t!='[CLS]' if t!='[SEP]']
    # cln_tokens

    # Convert to stanza library doc
    document = " ".join(cln_tokens)
    s_doc = nlp(document)

    # Remove punctuations
    rp_doc = [t for t in s_doc if t.dep_ != 'punct']
    # rp_doc

    # Remove stop-words
    rs1_doc = [t for t in rp_doc if t not in sv_stop_words.STOP_WORDS]
    rs_doc = [t for t in rs1_doc if not t.is_stop]
    # rp_doc

    # Remove number
    rn_doc = [t for t in rs_doc if t.pos_ != 'NUM']

    # Select only lemma and lower-case
    processed_sentence = " ".join([t.lemma_.lower() for t in rn_doc])
    return processed_sentence
```

```python
from tqdm import trange, tqdm
# from tqdm.notebook import tqdm
```

```python
tqdm().pandas()
news_items_pdf['ProcessedNews'] = news_items_pdf['TOKENS'].progress_apply(process_document) 
```

### Count vectorize

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import NMF, LatentDirichletAllocation, TruncatedSVD
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.manifold import TSNE
import numpy as np
```

```python
vectorizer = CountVectorizer(min_df=5, max_df=0.95, lowercase=True)
data_vectorized = vectorizer.fit_transform(news_items_pdf['ProcessedNews'])

print(f"Number if news items: {len(news_items_pdf['ProcessedNews'])}")
print("Output Vector Shape: ", data_vectorized.shape)
```

### Implement LDA

```python
NUM_TOPICS = 10
```

```python
# Latent Dirichlet Allocation Model
lda = LatentDirichletAllocation(n_components=NUM_TOPICS, max_iter=10, learning_method='online',verbose=True)
data_lda = lda.fit_transform(data_vectorized)
```

```python
# Functions for printing keywords for each topic
def selected_topics(model, vectorizer, top_n=10):
    for idx, topic in enumerate(model.components_):
        print("Topic %d:" % (idx))
        print([(vectorizer.get_feature_names()[i], topic[i])
                        for i in topic.argsort()[:-top_n - 1:-1]]) 
```

```python
# Keywords for topics clustered by Latent Dirichlet Allocation
print("LDA Model: \n*", " * "*10)
selected_topics(lda, vectorizer)
```

## Non-Negative Matrix Factorization Model

```python
# Non-Negative Matrix Factorization Model
nmf = NMF(n_components=NUM_TOPICS)  #  solver = 'mu'
data_nmf = nmf.fit_transform(data_vectorized) 
```

```python
# Keywords for topics clustered by Latent Semantic Indexing
print("NMF Model:\n*" ,' * ' * 10)
selected_topics(nmf, vectorizer)
```

## Latent Semantic Indexing Model

```python
# Latent Semantic Indexing Model using Truncated SVD
lsi = TruncatedSVD(n_components=NUM_TOPICS)
data_lsi = lsi.fit_transform(data_vectorized)
```

```python
# Keywords for topics clustered by Non-Negative Matrix Factorization
print("LSI Model:\n*" ,' * ' * 10)
selected_topics(lsi, vectorizer)
```
