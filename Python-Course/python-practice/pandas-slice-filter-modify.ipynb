```python
CSV_PATH = os.path.join("pandas-on-spyder" ,"TestData", "titanic.csv")
# Proper data loading
COLS_TO_USE = ['PassengerId','Survived', 'Pclass', 'Sex', 'Age', 'Fare', 'Cabin']
df = pd.read_csv(CSV_PATH, usecols=COLS_TO_USE, index_col='PassengerId')

df['Fare'].count()
df_prs = pd.to_numeric(df['Fare'], errors = "coerce")

# It has become a series
type(df_prs)
df_prs.count()
df_prs
# Apply iloc in a series
df_prs.iloc[260:270]
df_prs.iloc[263]
```

```python
test_df =  pd.DataFrame(np.random.rand(3,2), index = ['fst', 'sec', 'thd'], columns = ['cow', 'dog']) 
```

```python
# Filtering the rows with a list 
# i.e. [ True, True, True, True, True .... the number of row]
test_df[[True, Flase, True]]
```

```python
# Filtering the rows with a list from an expression
# test_df['cow'] > 0.5 givs a list of True/Flase of row size.
test_df[test_df['cow'] > 0.5]
```

```python
# Filter out how many children died.
df[(df['Age'] < 18) & (df['Survived'] == 0)]
```

```python
# Filter out how many children died - Using loc
df.loc[(df['Age']<18)&(df['Survived']==0),:]
```
