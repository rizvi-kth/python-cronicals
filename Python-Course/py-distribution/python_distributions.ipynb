A single number with mean 0.75 and 1 std-dev

```python
np.random.normal(0.75, 1)
```

```python
# Create 1000 random numbers with 0 mean and 2 std-dev and Sort
x_ax = [ np.random.normal(0, 2) for x in range(0, 1000)]
x_ax.sort()

# Calculate the probability of each number in a normal distribution with Probability-Density-Function(norm.pdf).
y_ax = norm.pdf(x_ax)

# plot with probability density function
plt.plot(x_ax, y_ax)
plt.show()
```

##### Kurtosis - how pointy the distribution head is?
- Positive value implies more pointy head distribution
- Negetive value implies more flat distribution
- value with 0 is considered normal distribution
- [Image of different Kurtosis](https://en.wikipedia.org/wiki/Kurtosis#/media/File:Standard_symmetric_pdfs.png)

```python
stats.kurtosis(x_ax)
```

##### Skew - how symetric the distribution is?
- Positive value implies right skew
- Negetive value implies left skew
- value within 0 - 0.5 is considered symetric
- value within 0.5 - 1 is considered little skewed
- value > 1 is considered very skewed

```python
stats.skew(x_ax)
```
